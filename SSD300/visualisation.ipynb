```python
with open("ap.json", "r") as file:
    ap = json.load(file)
classes = []
aps = []
for key, value in ap.items():
    classes.append(key)
    aps.append(value[-1])

plt.figure(dpi=300, figsize=(20, 10))
# creating the bar plot
plt.bar([i for i in range(2, 42, 2)], aps, width=1.5)
plt.xticks(fontsize=10)
 
# Create names on the x-axis
plt.xticks([i for i in range(2, 42, 2)], classes)

for i in range(2, 42, 2):
    plt.text(i - 0.65, aps[i//2-1]+0.005, f"{aps[i//2-1]:.4f}")
plt.ylim([0, 1])
plt.xlabel("Class")
plt.ylabel("AP")
plt.title("SSD300")
plt.show()
```
